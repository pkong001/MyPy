```python
list01 = [i.replace('"string",\n',"").replace('"double",\n',"").replace('"timestamp",\n',"").replace('"integer",',"").replace("\n","").strip()\
           for i in text01.split(":")]
```

```python
list01 = [i for i in text01.split(",")]

list_string = [i for i in list01 if "string" in i]
list_not_string = [i for i in list01 if ("string" not in i) & (len(i)>=3)]

list_string_clean = [i.replace('"string",\n',"").replace('"double",\n',"").replace('"timestamp",\n',"").replace('"integer",',"").replace("\n","")\
              .replace(': "string"',"").strip()\
              for i in list_string]  
list_not_string_clean = [i.replace('"integer"',"").replace('"double"',"").replace('"timestamp"',"").replace(':',"").replace("\n","")\
              .replace(': "string"',"").strip()\
              for i in list_not_string] 

list_agg = ["max("+i+").alias("+i+")" for i in list_not_string_clean]


text_string = ',\n '.join(list_string_clean)
text_agg = ',\n '.join(list_agg)



select_text = f"df_fact.groupBy({text_string}).agg({text_agg})"

print(select_text)
```
